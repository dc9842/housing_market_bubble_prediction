```python
import pandas as pd
```

```python
#create file index
years=["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019"]
file_idx=[]
for year in years:
    file_idx.append(year+"_houseEarningRatioData.csv")
```

```python
def load_data(filename): ##loads individual csv files
    df=pd.read_csv(filename)
    print("file length is:",len(df))
    df['Year']=filename[0:4]##year is parsed from first 4 char of filename
    return df, len(dfGodsize123!
                   
total_rows=0
for n,filename in enumerate(file_idx):
    if n==0:
        df_all_years, row_cnt=load_data(filename)
    else:       
        df_int, row_cnt=load_data(filename)
        df_all_years=df_all_years.append(df_int,ignore_index = True)   
    total_rows+=row_cnt
print("total rows:",total_rows)
#rename total column       
df_all_years.rename(columns={"Total:":"Average"},inplace=True)
#get column names from raw data and make lists to split data
cols1=list(df_all_years.columns[:7])
cols1.extend(list(df_all_years.columns[-1:]))

cols2=list(df_all_years.columns[7:])
cols2.extend([df_all_years.columns[0]])

df1=df_all_years[cols1]
df2=df_all_years[cols2]

df2=pd.melt(df2,id_vars=['RegionID','Year'],var_name='Job Description',value_name='HTI')

df1=df1.merge(df2,how='inner',on=['RegionID','Year'])
###load al table and covert year to string
df_alRatioData=pd.read_csv('alRatioData.csv')
df_alRatioData['year']=df_alRatioData['year'].astype(str)
df1=df1.merge(df_alRatioData,how='inner',left_on='Year',right_on='year')

df1.to_csv('bubblewarn.csv')
```
